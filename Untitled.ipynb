```python
from flirpy.camera.lepton import Lepton
from matplotlib import pyplot as plt
import numpy as np
%matplotlib inline
```

```python
#single capture

camera = Lepton()

img = camera.grab()
print(camera)
print(img)
plt.imshow(img, interpolation='nearest')
plt.show()

camera.close()
```

```python
camera = Lepton()

img = camera.grab(1)
print(camera)
print(img)
plt.imshow(img, interpolation='nearest')
plt.show()

camera.close()
```

```python
#pseudo-sync capture using a for loop
camList = np.array([1,3])
for i in camList:
    camera = Lepton(i)
    print(i)
    print(camera)

    img = camera.grab()
    #print(camera)
    #print(img)
    plt.imshow(img, interpolation='nearest')
    plt.show()
    
    camera.close()    
```
